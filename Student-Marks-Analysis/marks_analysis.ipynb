STUDENT MARK ANALYSIS

```python
import numpy as np
```

```python
# 5 Students and marks of 5 subjects
marks = np.array([
    [90,95,99,98,100],
    [70,80,90,68,78],
    [99,97,96,92,91],
    [85,86,89,90,88],
    [92,89,90,95,97]
])
```

Total Marks of Each Student

```python
total = np.sum(marks,axis=1)
print("Total Marks :- ", total)
```

Average Marks

```python
average_marks = np.average(marks, axis=1)
print("Average Marks :- ",average_marks)
```

Highest Scorer

```python
higher_scorer = np.argmax(total)
print("Highest Scorer Student :- ",higher_scorer)
```

Subject-wise Highest Marks

```python
subject_high = np.max(marks,axis=0)
print(subject_high)
```

Pass / Fail Analysis

```python
result = np.all(marks>=40, axis=1)
print("Pass Status:- ", result)
```

Grade Assignment

```python
grades = []
for avg in average_marks:
    if avg >= 90:
        grades.append("A")
    elif avg >= 75:
        grades.append("B")
    elif avg >= 60:
        grades.append("C")
    else:
        grades.append("D")
print("Grades :- ", grades)
```
