STUDENT MARKS ANALYSIS :- ADVANCED VERSION

```python
import numpy as np
import matplotlib.pyplot as plt
```

Generate Random Dataset

```python
num_student = 50
num_subject = 5
marks = np.random.randint(50,100, size=(num_student,num_subject))
subjects = ["Maths","Physics","Chemistry","Python","Numpy"]
print(marks)
```

Total, Average, Percentage

```python
total = np.sum(marks,axis=1)
average = np.mean(marks,axis=1)
percentage = (total/(num_subject*100)
)*100
print("Percentage :- \n",percentage)
```

Ranking System

```python
rank = np.argsort(-percentage) # sort in descending order
print("Top 5 Students :- ",rank[:5])
```

Grade System

```python
grades = np.where(percentage >= 90, "A",
                  np.where(percentage >= 75, "B",
                           np.where(percentage >= 60, "C",
                                    np.where(percentage >= 50, "D","F"))))
print("Grades :- ", grades)
```

Pass / Fail Analysis

```python
pass_status = np.all(marks >= 40, axis=1)
num_pass = np.sum(pass_status)
num_fail = num_student - num_pass
print("Pass Students :- ", num_pass)
print("Fail Students :- ", num_fail)
```

Subject-wise Analysis


-> Average per subject

```python
subject_average = np.mean(marks, axis=0)
for i in range(num_subject):
    print(subjects[i], "Average: ", subject_average[i])
```

Hardest Subject (Lowest Average)

```python
hardest_subj = subjects[np.argmin(subject_average)]
print("Hardest Subject : ", hardest_subj)
```

VISUALIZATION

Bar Chart :- Subject Average

```python
plt.bar(subjects , subject_average)
plt.title("Subject Wise Average Marks : ")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.show()
```

HISTOGRAM :- PERCENTAGE DISTRIBUTION

```python
plt.hist(percentage, bins=10)
plt.title("Percentage Distribution")
plt.xlabel("Percentage")
plt.ylabel("Number of Students")
plt.show()
```

SAVE RESULT TO CSV

```python
result = np.column_stack((marks, total, percentage))
np.savetxt("Student_results.csv",
           result, delimiter=",", fmt='%0.2f')
```
