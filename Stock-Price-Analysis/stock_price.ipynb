STOCK PRICE ANALYSIS

```python
import numpy as np
```

```python
prices = np.array([100, 101, 102, 103, 104])
```

DAILY RETURNS

```python
daily_returns = (prices[1:]-prices[:-1])/prices[:-1]
print("Daily Return Prices :- ",daily_returns)
```

Average Return

```python
avg_return = np.mean(daily_returns)
print("Average Daily Return :- ",avg_return)
```

Volatility(Risk)
Standard Deviation of Returns

```python
volatility = np.std(daily_returns)
print("Volatility (Risk) :- ",volatility)
```

Best and Worst Day

```python
best_day = np.argmax(daily_returns)
print("Best Day Index :- ",best_day)
worst_day = np.argmin(daily_returns)
print("Worst Day Index :- ",worst_day)
```

Moving Average (Trend Detection)

```python
window = 3
moving_avg = np.convolve(prices, np.ones(window)/window, mode='valid')
print("Moving Average :- ",moving_avg)
```

Cumulative Return

```python
cum_return = np.cumsum(prices)
print(cum_return)
```

```python
cumulative_return = (prices[-1] - prices[0]) / prices[0]
print("Cumulative Return:", cumulative_return)
```

Sharpe ratio (Risk-adjust return)

```python
sharpe_ratio = avg_return/volatility
print("Sharpe Ratio :- ",sharpe_ratio)
```

Detect Bull and Bear Days

```python
bull_day = daily_returns[daily_returns > 0]
bear_day = daily_returns[daily_returns < 0]
print("Bull Day :- ",bull_day)
print("Bear Day :- ",bear_day)
```

Highest & Lowest Prices

```python
print("Highest Price:", np.max(prices))
print("Lowest Price:", np.min(prices))
```
